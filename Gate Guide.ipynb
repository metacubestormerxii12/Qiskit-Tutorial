```python
import qiskit
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
```

```python
c = QuantumCircuit(2,2)
c.h(0)

statevec = execute(c, backend = Aer.get_backend("statevector_simulator")).result().get_statevector()
```

```python
plot_bloch_multivector(statevec)
```
